```python
from pathlib import Path
import json
from collections import defaultdict

CHUNK_DIR = Path("data/chunks")
CHUNK_DIR.mkdir(parents=True, exist_ok=True)

grouped_chunks = defaultdict(list)

for chunk in chunks:
    title = chunk.metadata["title"]
    grouped_chunks[title].append(chunk)

for title, doc_chunks in grouped_chunks.items():

    chunk_list = []

    doc_id = doc_chunks[0].metadata["doc_id"]

    for idx, chunk in enumerate(doc_chunks):

        chunk_list.append(
            {
                "chunk_id": f"{doc_id}_{idx}",
                "content": chunk.page_content,
                "chunk_length": len(chunk.page_content),
                "metadata": chunk.metadata
            }
        )

    output_file = (
        CHUNK_DIR /
        f"{title.replace(' ', '_')}_chunks.json"
    )

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(
            chunk_list,
            f,
            indent=4,
            ensure_ascii=False
        )

print(f"Saved {len(grouped_chunks)} files")
```

```python
ROUTING_DATASET = [

    {
        "question": "What is fault tolerance?",
        "expected_route": "rag"
    },

    {
        "question": "What is CAP theorem?",
        "expected_route": "rag"
    },

    {
        "question": "Difference between monolith and microservices?",
        "expected_route": "rag"
    },

    {
        "question": "What is circuit breaker pattern?",
        "expected_route": "rag"
    },

    {
        "question": "Who is Elon Musk?",
        "expected_route": "direct"
    },

    {
        "question": "Capital of India?",
        "expected_route": "direct"
    },

    {
        "question": "Who is Virat Kohli?",
        "expected_route": "direct"
    },

    {
        "question": "What is Machine Learning?",
        "expected_route": "direct"
    }

]
```

```python
correct = 0

total = len(
    ROUTING_DATASET
)

for item in ROUTING_DATASET:

    question = item["question"]

    expected_route = item["expected_route"]

    result = graph.invoke(
        {
            "question": question,
            "retriever": retriever,
            "llm": llm
        }
    )

    predicted_route = result["route"]

    is_correct = (
        predicted_route == expected_route
    )

    if is_correct:
        correct += 1

    print("\n" + "=" * 70)

    print(
        f"Question: {question}"
    )

    print(
        f"Expected Route: {expected_route}"
    )

    print(
        f"Predicted Route: {predicted_route}"
    )

    print(
        f"Correct: {is_correct}"
    )
```

```python
accuracy = (
    correct / total
) * 100

print("\n" + "=" * 70)

print(
    f"Total Questions: {total}"
)

print(
    f"Correct Routes: {correct}"
)

print(
    f"Routing Accuracy: {accuracy:.2f}%"
)
```
